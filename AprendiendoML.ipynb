## Ejemplo de Machine Learning

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
%matplotlib inline
```

```python
np.random.seed(5)

# Iris dataset (Ronald Fisher - 1936)
iris = datasets.load_iris()
```

```python
X = iris.data
X[:10]
# Largo de sépalo, Ancho de sépalo, Largo de pétalo, Ancho de pétalo
```

```python
y = iris.target
y
# setosa, versicolara, virginica
```

```python
estimator = KMeans(n_clusters=3)
estimator.fit(X)

labels = estimator.labels_
labels
```

```python
estimated_fig = plt.figure()

ax = Axes3D(estimated_fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=labels.astype(np.float))

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Petal width')
ax.set_ylabel('Sepal length')
ax.set_zlabel('Petal length')

plt.show()
```

```python
real_fig = plt.figure()

ax = Axes3D(real_fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)

for name, label in [('Setosa', 0),
                    ('Versicolour', 1),
                    ('Virginica', 2)]:
    ax.text3D(X[y == label, 3].mean(),
              X[y == label, 0].mean() + 1.5,
              X[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
              
ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=y)

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Petal width')
ax.set_ylabel('Sepal length')
ax.set_zlabel('Petal length')

plt.show()
```
